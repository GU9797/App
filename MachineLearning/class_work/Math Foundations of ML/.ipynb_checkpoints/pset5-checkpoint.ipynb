a)
Overall, the classifier trained was mediocre.
The accuracy is 0.6 if we use a classification rule where positive values are 1 and negative values are -1.
Also, the best rated joke is not the same between the predicted ratings and the true ratings.

```python
jester = loadmat('jesterdata.mat')
newuser = loadmat('newuser.mat')
```

```python
yindices = [i for i, rating in enumerate(newuser['y']) if rating != -99]
ytestindices = [i for i, rating in enumerate(newuser['y']) if rating == -99]
x = [jester['X'][j][:20] for j in yindices]
y = [rating for rating in newuser['y'] if rating != -99]
w = np.matmul(np.matmul(np.linalg.inv(np.matmul(np.transpose(x), x)), np.transpose(x)), y)
xtest = [jester['X'][j][:20] for j in ytestindices]
ytest = [newuser['truey'][k] for k in ytestindices]
ypred = np.matmul(xtest, w)
acc_vec = ypred * ytest
accuracy = len([n for n in acc_vec if n > 0]) / len(ytest)
print(accuracy)
print(list(ypred).index(max(ypred)))
print(list(ytest).index(max(ytest)))
```

b)
I plan on using the singular value decomposition approach to underdetermined least squares, where the weight matrix is minimized using $$ w = V\Sigma^{+}U^{T}y $$ . It performs better than the first method using the same classification rule, with an accuracy of 0.787. However, the best joke still does not match. 

```python
xtrain = [jester['X'][j] for j in yindices]
xtest = [jester['X'][j] for j in ytestindices]
A = np.array(xtrain)
# Singular-value decomposition
U, s, VT = svd(A)
# create m x n Sigma matrix
Sigma = np.zeros((A.shape[0], A.shape[1]))
# populate Sigma with n x n diagonal matrix
Sigma[:A.shape[0], :A.shape[0]] = np.diag(s)
w = np.matmul(np.matmul((np.matmul(np.transpose(VT), pinv(Sigma))), np.transpose(U)), y)
ypred = np.matmul(xtest, w)
acc_vec = ypred * ytest
accuracy = len([n for n in acc_vec if n > 0]) / len(ytest)
print(accuracy)
print(list(ypred).index(max(ypred)))
print(list(ytest).index(max(ytest)))
```

f)
The power method essentially multiplies a random vector by diagonalizable matrix A successively until the final vector reflects the largest eigenvalue of A. Normalization is done to prevent the vector from growing infinitely or shrinking to 0 and because we are not interested in the length of the vector, but rather its direction. The power_iteration function and numpy's svd function both return the same dominant eigenvector.

```python
def power_iteration(A, num_simulations):
    # Ideally choose a random vector
    # To decrease the chance that our vector
    # Is orthogonal to the eigenvector
    b_k = np.random.rand(A.shape[1]).reshape(-1,1)

    for _ in range(num_simulations):
        # calculate the matrix-by-vector product Ab
        b_k1 = np.dot(np.transpose(A), b_k)

        # calculate the norm
        b_k1_norm = np.linalg.norm(b_k1)

        # re normalize the vector
        b_k = b_k1 / b_k1_norm
        
    sigma = np.linalg.norm(np.matmul(A, b_k))
    u = np.matmul(A, b_k) / sigma
    return b_k, sigma, u
```

```python
v, sigma, poweru = power_iteration(np.matmul(np.transpose(jester['X']), jester['X']), 100)
svdu, s, VT = svd(np.matmul(np.transpose(jester['X']), jester['X']))
print(poweru)
print("--------------------")
print(np.transpose(svdu)[0])
```

A vector that will not converge is a vector orthogonal to the dominant eigenvector. An example is vector "ortho" of same length as the eigenvector that contains 1's for its entire length, with the exception of the last value, which is calculated to be the negative of the sum of the preceding values in the eigenvector divided by the last value in the eigenvector. This results in the dot product being 0 (or as close to it as possible given underflow).

```python
face_emotion = loadmat('face_emotion_data.mat')
x = face_emotion['X']
y = face_emotion['y']
```

```python

def crossval(x, y, cv, reg_values, function):
    xrows = len(x)
    cvsplit = int(xrows / cv)
    errors = []
    xslices, yslices = [], []
    for i in range(0, int(xrows), cvsplit):
        xslices.append(x[i:i + cvsplit])
        yslices.append(y[i:i + cvsplit])
    error_vec = []
    for k in range(cv):
        xregholdout = xslices[k]
        yregholdout = yslices[k]
        for j in range(cv):
            if j != k:
                xtest = xslices[j]
                ytest = yslices[j]
                xtrain = np.array([piece for n, piece in enumerate(xslices) if (n != j) and (n != k)]).reshape(6 * cvsplit, x.shape[1])
                ytrain = np.array([piece for n, piece in enumerate(yslices) if (n != j) and (n != k)]).reshape(6 * cvsplit, 1)
                best_acc, best_reg = 0, 0
                for p in reg_values:
                    w = function(xtrain, ytrain, p)
                    
                    ypredholdout = np.matmul(xregholdout, w)
                    acc_vec = ypredholdout * yregholdout
                    accuracy = len([n for n in acc_vec if n > 0]) / len(yregholdout)
                    if accuracy > best_acc:
                        best_acc = accuracy
                        best_reg = p
                w = function(xtrain, ytrain, best_reg)
                ypred = np.matmul(xtest, w)
                acc_vec = ypred * ytest
                accuracy = len([n for n in acc_vec if n > 0]) / len(ytest)
                error_vec.append(accuracy)
    return np.average(error_vec)

            
def truncate(x, y, k):
    A = np.array(x)
    #print(A.shape)
    # Singular-value decomposition
    U, s, VT = svd(A)
    # create m x n Sigma matrix
    Sigma = np.zeros((A.shape[0], A.shape[1]))
    # populate Sigma with n x n diagonal matrix
    svs = (1 / np.diag(s))[:k]
    Sigma[:A.shape[0], :A.shape[0]][:svs.shape[0], :svs.shape[1]] = svs
    Sigma[Sigma == np.inf] = 0
    w = np.matmul(np.matmul(np.transpose(VT), np.matmul(np.transpose(Sigma), np.transpose(U))), y)
    return w
    

def reg_ls(x, y, k):
    A = np.array(x)
    #print(A.shape)
    # Singular-value decomposition
    U, s, VT = svd(A)
    # create m x n Sigma matrix
    Sigma = np.zeros((A.shape[0], A.shape[1]))
    # populate Sigma with n x n diagonal matrix
    Sigma[:A.shape[0], :A.shape[0]] = diag(s)
    #Sigma[Sigma == np.inf] = 0
    ridge_inv = np.linalg.inv(np.add(np.matmul(np.transpose(Sigma), Sigma), k * np.identity(x.shape[1])))
    w = np.matmul(np.matmul(np.transpose(VT), np.matmul(np.matmul(ridge_inv, np.transpose(Sigma)), np.transpose(U))), y)
    
    #w = np.matmul(np.matmul(np.linalg.inv(np.add(np.matmul(np.transpose(x), x), k * np.identity(x.shape[1]))), np.transpose(x)), y)
    return w
 
def lsquares(x, y):
    w = np.matmul(np.matmul(np.linalg.inv(np.matmul(np.transpose(x), x)), np.transpose(x)), y)
    return w
```

```python
## deblurring


n = 500
k = 30
sigma = 0.01

# generate random piecewise constant signal
w = np.zeros((n, 1))
w[0] = np.random.standard_normal()
for i in range(1, n):
    if np.random.rand(1) < 0.95:
        w[i] = w[i-1]
    else:
        w[i] = np.random.standard_normal()


# generate k-point averaging function
h = np.ones(k) / k

# make a matrix for blurring
m = n + k - 1
X = np.zeros((m, m))
for i in range(m):
    if i < k:
        X[i, :i+1] = h[:i+1]
    else:
        X[i, i - k: i] = h

X = X[:, 0:n]

# blurred signal + noise
y = np.dot(X, w) + sigma*np.random.standard_normal(size=(m, 1))

# plot
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.set_title('signal')
ax1.plot(w)
ax2.set_title('blurred and noisy version')
ax2.plot(y[0:n])

plt.show()
```
